```python
cap_clean = cap[['S. No.', 'Category', 'State/UT', 'Number of Cases - Registered',
       'Number of Cases - Quashed/Stayed by Courts']]
cap_clean.head()
```

```python
efp = pd.read_csv("2018 Escapes from Police Custody.csv")
efp.head()
```

```python
efp.columns
```

```python
efp_clean = efp[['S. No.', 'Cases Reported','Persons Escaped - From lockup','Persons Escaped - Outside Lockup',
       'Persons Escaped - Total']]
efp_clean.head()
```

```python
vor = pd.read_csv("2018 Victims of Rape.csv")
vor.head()
```

```python
vor.columns
```

```python
vor_clean = vor[['S. No','Cases Reported','Child Victims of Rape (Below 18 Yrs) - Below 6 Years',
       'Child Victims of Rape (Below 18 Yrs) - 6 Years & Above ',
       'Child Victims of Rape (Below 18 Yrs) - 12 Years & Above ',]]
vor_clean.head()
```

```python
def CreateDatabase():
    conn = psycopg2.connect("host=localhost dbname= postgres user=postgres password=root")
    cur = conn.cursor()
    conn.set_session(autocommit=True)
    
    cur.execute("drop database if exists crimes")
    cur.execute("create database crimes")
    
    conn.close()
    
    conn = psycopg2.connect("host=localhost dbname=crimes user=postgres password=root")
    cur = conn.cursor()
    conn.set_session(autocommit=True)
    return cur,conn    
```

```python
cur, conn = CreateDatabase()
```

```python
cases = ("""create table if not exists cases(SNo integer primary key,
Category varchar,
State varchar,
NumberofCasesRegistered integer,
NumberofCasesStayedbyCourts integer)""")
```

```python
# cur.execute("rollback")
# conn.commit()
```

```python
cur.execute(cases)
```

```python
escapes = ("""create table if not exists escapes(SNo integer,
CasesReported integer,
PersonsEscapedFromlockup  integer,
PersonsEscapedOutsideLockup integer,
PersonsEscapedTotal integer)""")
```

```python
cur.execute(escapes)
```

```python
victims = ("""create table if not exists victims(SNo integer,
CasesReported integer,
ChildVictimsofRapeBelow6Years integer,
ChildVictimsofRape6YearsAbove integer,
ChildVictimsofRape12YearsAbove integer)""")
```

```python
cur.execute(victims)
```

```python
cases_insert = ("""insert into cases
(SNo,Category,State,NumberofCasesRegistered,NumberofCasesStayedbyCourts)
VALUES (%s,%s,%s,%s,%s)
""")
```

```python
for i, row in cap_clean.iterrows():
    cur.execute(cases_insert,list(row))
```

```python
escapes_insert = ("""insert into escapes
(SNo,CasesReported,PersonsEscapedFromlockup,PersonsEscapedOutsideLockup,PersonsEscapedTotal)
VALUES (%s,%s,%s,%s,%s)
""")
```

```python
for i, row in efp_clean.iterrows():
    cur.execute(escapes_insert,list(row))
```

```python
victims_insert = ("""insert into victims
(SNo,CasesReported,ChildVictimsofRapeBelow6Years,ChildVictimsofRape6YearsAbove,ChildVictimsofRape12YearsAbove)
VALUES (%s,%s,%s,%s,%s)
""")
```

```python
for i, row in vor_clean.iterrows():
    cur.execute(victims_insert,list(row))
```
